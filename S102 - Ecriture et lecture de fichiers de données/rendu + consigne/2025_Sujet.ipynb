### T√¢che 3 :

Pour cette t√¢che de mise en place d'un rapport d'√©tudes statistique, vous pouvez utiliser les donn√©es du fichier `pays_zones.csv` qui associe chaque pays √† une macro-zone g√©ographique europ√©enne : 
- Europe m√©diterran√©enne

- Europe de l‚Äôest

Il est obligatoire de produire une statistique de la r√©partition des volontaires par macro-zone g√©ographique.
En plus, vous devez choisir au moins une autre variable qui vous para√Æt pertinente (exemples : GENRE, TITRE_ETUDE, tranches d‚Äô√¢ge, etc.).

Le rapport sera enregistr√© dans un fichier texte nomm√© `rapport_stat.txt`.

*La forme du rapport et le choix du ou des indicateurs est laiss√© √† votre jugement.*

#### Impl√©mentation

Vous devez √©crire le code de cette t√¢che dans cette section du document.

```python


# CHARGEMENT DES ZONES 
liste_association_pays_zone = []

f_ref = open("pays_zones.csv", "r", encoding="utf-8")
lignes_ref = f_ref.readlines()
f_ref.close()

# Lecture du fichier pays_zones.csv (on saute l'ent√™te)
for k in range(1, len(lignes_ref)):
    ligne_net = lignes_ref[k].strip()
    if ligne_net != "":
        morceaux = ligne_net.split(";")
        pays_lu = morceaux[0]
        zone_lue = morceaux[1]
        
        # On ajoute la paire [Pays, Zone] dans notre liste
        liste_association_pays_zone.append([pays_lu, zone_lue])


# LECTURE DES VOLONTAIRES ET COMPTAGE
f_donnees = open("base_volontaires.csv", "r", encoding="utf-8")
lignes_volontaires = f_donnees.readlines()
f_donnees.close()

# Listes pour le comptage : [ ["Nom", Nombre], ... ]
stockage_comptage_zones = []
stockage_comptage_etudes = []

total_lignes_traitees = 0

for j in range(1, len(lignes_volontaires)):
    une_ligne = lignes_volontaires[j].strip()
    
    if une_ligne != "":
        total_lignes_traitees = total_lignes_traitees + 1
        
        infos = une_ligne.split(";")
        
        # Colonne 5 = PAYS, Colonne 6 = TITRE_ETUDE
        pays_candidat = infos[5]
        etude_candidat = infos[6]
        
        # Recherche de la Zone 
        zone_trouvee = "Zone Inconnue"
        
        # Parcours de la liste de correspondance
        for couple in liste_association_pays_zone:
            if couple[0] == pays_candidat:
                zone_trouvee = couple[1]
                
        
        #  Comptage de la Zone 
        deja_present_zone = False
        
        for element_zone in stockage_comptage_zones:
            if element_zone[0] == zone_trouvee:
                element_zone[1] = element_zone[1] + 1
                deja_present_zone = True
               
        
        if deja_present_zone == False:
            stockage_comptage_zones.append([zone_trouvee, 1])

        #  Comptage des Etudes 
        if etude_candidat == "null" or etude_candidat == "":
            etude_candidat = "Non renseigne"
            
        deja_present_etude = False
        
        for element_etude in stockage_comptage_etudes:
            if element_etude[0] == etude_candidat:
                element_etude[1] = element_etude[1] + 1
                deja_present_etude = True
            
                
        if deja_present_etude == False:
            stockage_comptage_etudes.append([etude_candidat, 1])


 #ECRITURE DU RAPPORT
f_final = open("rapport_stat.txt", "w", encoding="utf-8")

f_final.write("##################################################\n")
f_final.write("#   ANALYSE DES PROFILS - PROGRAMME MARS 2035    #\n")
f_final.write("##################################################\n\n")

f_final.write("Nombre de candidats traites : " + str(total_lignes_traitees) + "\n\n")

# Partie Zones
f_final.write(">>> PARTIE 1 : ORIGINE GEOGRAPHIQUE <<<\n")
f_final.write("Objectif : Assurer la diversite geographique.\n\n")

for item in stockage_comptage_zones:
    nom_z = item[0]
    nb_z = item[1]
    pct = (nb_z / total_lignes_traitees) * 100
    ligne_txt = " [ZONE] " + nom_z + " .................... " + str(nb_z) + " (" + str(int(pct)) + "%)\n"
    f_final.write(ligne_txt)

f_final.write("\n")

# Partie √âtudes
f_final.write(">>> PARTIE 2 : NIVEAUX D'ETUDES <<<\n")
f_final.write("Objectif : Verifier les qualifications.\n\n")

for item in stockage_comptage_etudes:
    nom_e = item[0]
    nb_e = item[1]
    pct = (nb_e / total_lignes_traitees) * 100
    ligne_txt = " [DIPLOME] " + nom_e + " .................... " + str(nb_e) + " (" + str(int(pct)) + "%)\n"
    f_final.write(ligne_txt)

f_final.close()

print("Termin√© !")
```
